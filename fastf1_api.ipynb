```python
import fastf1
import pandas as pd
from google.colab import files

# Danh sách để lưu các DataFrame cho từng năm
all_years_data = {}

# Lặp qua các năm từ 2019 đến 2026
for year in range(2019, 2026):
    all_data_laps = []
    all_data_results = []
    all_data_weather = []
    print(f"Tải dữ liệu năm: {year}")

    try:
        schedule = fastf1.get_event_schedule(year)
        unique_locations = schedule['Location'].unique()
        for loc in unique_locations:
            try:
                session = fastf1.get_session(year, loc, 'Q')
                session2 = fastf1.get_session(year, loc, 'R')
                session.load()
                session2.load()

                # Lấy dữ liệu vòng đua cho phiên Q
                laps = session.laps.copy()
                laps['Location'] = loc  # Thêm cột để biết nguồn dữ liệu
                laps['session'] = "Q"
                laps['year'] = year

                # Lấy dữ liệu vòng đua cho phiên R
                laps2 = session2.laps.copy()
                laps2['Location'] = loc  # Thêm cột để biết nguồn dữ liệu
                laps2['session'] = "R"
                laps2['year'] = year

                # Lấy dữ liệu kết quả cho phiên Q
                results = session.results.copy()
                results['Location'] = loc
                results['session'] = "Q"
                results['year'] = year

                # Lấy dữ liệu kết quả cho phiên R
                results2 = session2.results.copy()
                results2['Location'] = loc
                results2['session'] = "R"
                results2['year'] = year

                # Lấy dữ liệu thời tiết cho phiên Q
                weather_data = session.weather_data.copy()
                weather_data['Location'] = loc
                weather_data['session'] = "Q"
                weather_data['year'] = year

                # Lấy dữ liệu thời tiết cho phiên R
                weather_data2 = session2.weather_data.copy()
                weather_data2['Location'] = loc
                weather_data2['session'] = "R"
                weather_data2['year'] = year

                # Gộp tất cả dữ liệu vào một DataFrame
                all_data_laps.append(laps)
                all_data_laps.append(laps2)
                all_data_results.append(results)
                all_data_results.append(results2)
                all_data_weather.append(weather_data)
                all_data_weather.append(weather_data2)

            except Exception as e:
                print(f"Lỗi khi tải dữ liệu cho địa điểm {loc} trong năm {year}: {e}")

        # Gộp lại tất cả các DataFrame cho năm hiện tại
        combined_df_laps = pd.concat(all_data_laps, ignore_index=True)
        combined_df_results = pd.concat(all_data_results, ignore_index=True)
        combined_df_weather = pd.concat(all_data_weather, ignore_index=True)

        # Lưu DataFrame vào file CSV
        combined_df_laps.to_csv(f"f1_laps_{year}.csv", index=False)
        combined_df_results.to_csv(f"f1_results_{year}.csv", index=False)
        combined_df_weather.to_csv(f"f1_weather_{year}.csv", index=False)

        files.download(f"f1_laps_{year}.csv")
        files.download(f"f1_results_{year}.csv")
        files.download(f"f1_weather_{year}.csv")

        print(f"Dữ liệu cho năm {year} đã được lưu vào các file f1_laps_{year}.csv, f1_results_{year}.csv, và f1_weather_{year}.csv.")

    except Exception as e:
        print(f"Lỗi khi tải dữ liệu cho năm {year}: {e}")

print("Quá trình thu thập dữ liệu hoàn tất.")
```
